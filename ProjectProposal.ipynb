```python
import pandas as pd
import numpy as np
```

```python
covid = pd.read_csv('US Statewise COVID.csv',header=0,
         names=['country_region_code','country_region','sub_region','iso_code','date','retail_and_recreation','grocery_and_pharmacy','parks','transit_stations','workplaces','residential'])
```

```python
srp = covid.groupby("sub_region")["parks"].mean()
srp
```
